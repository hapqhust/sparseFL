```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./dataset_idx/cifar10/sparse_dir0.1_U100/100client/cifar10_sparse_stat.csv", header=None)
df.head(10)
```

```python
num_zeros = (df == 0).sum().sum()
print(num_zeros)
total_sample = len(df)*len(df.T)
print(num_zeros/total_sample)
```

```python
total_sample = df.sum().sum()
print(total_sample)

client_samples = df.T.sum().to_list()
print(client_samples)
```

```python
thresold1 = 50
thresold2 = total_sample/len(df)

print(thresold2)

scarcity_T = 0
scarcity_M = 0

for client_sample in client_samples:
    if client_sample <= thresold1:
        scarcity_T += 1
    if client_sample <= thresold2:
        scarcity_M += 1

print(scarcity_T/len(df))
print(scarcity_M/len(df))
```
